```python
import json
import os
tags_dict = {1: "divorce/tags.txt", 2: "labor/tags.txt", 3: "loan/tags.txt"}
data_type_dict = {1: "divorce", 2: "labor", 3: "loan"}
output_dict = {1: "divorce/data.txt", 2: "labor/data.txt", 3: "loan/data.txt"}
```

```python
def extract_tags_dict(tags_file):
    # 构建标签字典
    tags = {}
    for tag in tags_file.readlines():
        tags[tag.strip()] = len(tags)
    return tags
```

```python
def extract_train_data(tags_file, data_type, output_file):
    tags = extract_tags_dict(tags_file)
    data_file = open(os.path.join(data_type,"data_small_selected.json"),'r',encoding='utf-8')
    data_file2 = open(os.path.join(data_type,"train_selected.json"),'r',encoding='utf-8')
    text=dict()
    j = 1
    for line in data_file.readlines():
        data = json.loads(line, encoding='utf-8')
        for dic in data:
            label = [0] * len(tags)
            sentence = dic['sentence'].strip().replace('\t',' ').replace("\n",'')
            if text.get(sentence) ==None:
                text[sentence]=1
            else:
                continue
            print(data_type, j)
            for l in dic['labels']:
                label[tags[l]] = 1
            output_file.write(str(j) + '\t' + sentence + '\t' + " ".join(list(map(str, label))) + '\n')
            j += 1
    for line in data_file2.readlines():
        data = json.loads(line, encoding='utf-8')
        for dic in data:
            label = [0] * len(tags)
            sentence = dic['sentence'].strip().replace('\t',' ').replace('\n','')
            if text.get(sentence) ==None:
                text[sentence]=1
            else:
                continue
            print(data_type, j)
            for l in dic['labels']:
                label[tags[l]] = 1
            output_file.write(str(j) + '\t' + sentence + '\t' + " ".join(list(map(str, label))) + '\n')
            j += 1
    output_file.close()
    data_file.close()
    data_file2.close()
    tags_file.close()
```

```python
if __name__ == "__main__":
    for i in range(1, 4):
        tags_input_file = open(tags_dict[i], encoding='utf-8')
        data_output_file = open(output_dict[i], 'w', encoding='utf-8')
        extract_train_data(tags_input_file,data_type_dict[i],data_output_file)
```
